```python
import pandas as pd
import re
from collections import Counter
import heapq

def split_file(log_file_path, chunk_size):
    df = pd.read_excel(log_file_path)  
    chunks = [df.iloc[i:i + chunk_size] for i in range(0, len(df), chunk_size)]
    return chunks

def count_errors(df_chunk,error_dict):

    for index, row in df_chunk.iterrows():
        log_message = str(row[0])  # התאמת העמודה לפי הצורך
        error_match = re.search(r"ERR_(\d+)", log_message)

        if error_match:
            error_code = error_match.group(1)    
            error_dict[error_code] = error_dict.get(error_code, 0) + 1   

def get_top_n_errors(error_dict, N):
    return heapq.nlargest(N, error_dict.items(), key=lambda x: x[1])


def process_logs(log_file_path, chunk_size, N):
    chunks = split_file(log_file_path, chunk_size)
    error_dict = {}
    [count_errors(chunk,error_dict) for chunk in chunks]
    return get_top_n_errors(error_dict, N)

log_file_path = r"logs.txt.xlsx"  
chunk_size = 1000   
N = 2  

result = process_logs(log_file_path, chunk_size, N)
print(" קודי השגיאה השכיחים ביותר:", result)
```

```python
import pandas as pd

def split_file(log_file_path, chunk_size):
    # קריאת הקובץ
    df = pd.read_excel(log_file_path)
    
    # חלוקה לפי chunk_size
    chunks = [df.iloc[i:i + chunk_size] for i in range(0, len(df), chunk_size)]
    
    return chunks

def calculate_hourly_avg_per_chunk(chunks):
    hourly_avgs = []
    
    for chunk in chunks:
        # המרת timestamp לפורמט תאריך
        chunk['timestamp'] = pd.to_datetime(chunk['timestamp'], errors='coerce', dayfirst=True)
        
        # הסרת שורות עם תאריכים לא חוקיים (NaT)
        chunk = chunk.dropna(subset=['timestamp'])
        
        # יצירת עמודה של שעה עגולה
        chunk['hour'] = chunk['timestamp'].dt.floor('H') 
        
        # חישוב ממוצע שעתיים עבור כל קבוצה
        hourly_avg = chunk.groupby('hour', as_index=False)['value'].mean()
        hourly_avgs.append(hourly_avg)
    
    return hourly_avgs

def combine_hourly_avgs(hourly_avgs):
    # איחוד כל התוצאות לקובץ אחד
    combined_df = pd.concat(hourly_avgs, ignore_index=True)
    
    # סידור מחדש לפי timestamp (שעה)
    combined_df = combined_df.sort_values(by='hour')
    
    return combined_df

def process_log_file(log_file_path, chunk_size):
    # חלוקה לקבוצות
    chunks = split_file(log_file_path, chunk_size)
    
    # חישוב ממוצע שעתיים עבור כל חלק
    hourly_avgs = calculate_hourly_avg_per_chunk(chunks)
    
    # איחוד כל התוצאות
    final_df = combine_hourly_avgs(hourly_avgs)
    
    # שמירת התוצאה לקובץ סופי
    final_df.to_csv('final_hourly_avg.csv', index=False)
    print("The final hourly averages have been saved to 'final_hourly_avg.csv'.")

# קריאה לפונקציה עם נתיב הקובץ
log_file_path = r"C:\Users\user\time_series.xlsx"
chunk_size = 1000  # לדוגמה, כל 1000 שורות יהיו חלק נפרד
process_log_file(log_file_path, chunk_size)
```
